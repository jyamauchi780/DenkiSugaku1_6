# 解析解の描画

```python
import numpy as np
import matplotlib.pyplot as plt

Delta_x = 0.01
a = 0.0; b = 5.0
x = np.arange(a, b, Delta_x)
y0 = 1.0           # 初期条件 y0 = y(0) この行を修正
y = y0 * np.exp(x) # 解析解 この行を修正

fig, ax = plt.subplots()
plt.rcParams["font.size"] = 14
ax.plot(x, y, lw=4, label='Analytical solution')
ax.set_xlabel("x"); ax.set_ylabel("y")
ax.grid(linestyle=':')
plt.legend(); plt.show()
```

# 微分方程式の定義

```python
def f(x, y):
    return y # この行を修正
```

# オイラー法による数値解の導出

```python
h = 0.5
x_euler = np.arange(a, b, h)
y_euler = np.zeros(len(x_euler))
y_euler[0] = y0
for i in range(len(x_euler)-1):
    y_euler[i+1] \
        = y_euler[i] + h*f(x_euler[i], y_euler[i])

fig, ax = plt.subplots()
plt.rcParams["font.size"] = 14
ax.plot(x, y, c='k', lw=4, label='Analytical solution')
ax.scatter(x_euler, y_euler, s=80, marker='o', fc='none', ec='b', lw=2, label='Euler method')
ax.set_xlabel("x"); ax.set_ylabel("y")
ax.grid(linestyle=':')
plt.legend(); plt.show()
```

# 改良オイラー法による数値解の導出

```python
h = 0.5
x_euler2 = np.arange(a, b, h)
y_euler2 = np.zeros(len(x_euler2))
y_euler2[0] = y0
for i in range(len(x_euler2)-1):
    y_euler2[i+1] \
    = y_euler2[i] + (h/2)*(f(x_euler2[i], y_euler2[i]) \
           + f(x_euler2[i+1], y_euler2[i]+h*f(x_euler2[i], y_euler2[i])))

fig, ax = plt.subplots()
plt.rcParams["font.size"] = 14
ax.plot(x, y, c='k', lw=4, label='Analytical solution')
ax.scatter(x_euler, y_euler, s=80, marker='o', fc='none', ec='b', lw=2, label='Euler method')
ax.scatter(x_euler2, y_euler2, s=80, marker='o', fc='none', ec='g', lw=2, label='Improved Euler method')
ax.set_xlabel("x"); ax.set_ylabel("y")
ax.grid(linestyle=':')
plt.legend(); plt.show()
```
